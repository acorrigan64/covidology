```python
import requests
import json
import pandas as pd
import tqdm.notebook as tqdm
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import twint
import nest_asyncio
# from datetime import datetime
import datetime

# Used to fix threading error from twint
nest_asyncio.apply()
```

```python
c = twint.Config()
```

```python
c.Search = 'coronavirus OR COVID-19'
c.Near = 'Bristol'
c.Location = True
c.Limit = 20
c.Lang = 'en'
c.Store_csv = True
c.Show_hashtags = True
c.Count = True
c.Filter_retweets = True

c.Since = '2020-03-20'
c.Until = '2020-03-21'

c.Output = "tweets.csv"

c.Verified = False
c.Popular_tweets = True
```

```python
# Test run
twint.run.Search(c)
```

```python
# Get list of required cities and dates to pull tweets from
cities = ['London', 'Bristol', 'Manchester', 'Liverpool', 'Birmingham', 'Leeds', 'Oxford', 'Glasgow', 'Dublin', 'Cardiff']
dates = pd.date_range(start="2019-12-25",end="2020-03-25").tolist()
```

```python
# Fetch 100 tweets for each city in date range 25-12-2019 to 25-03-2020
c.Limit = 100
c.Hide_output = True

for city in tqdm.tqdm(cities, leave=False):
    c.Near = city
    
    for date in tqdm.tqdm(dates):
        c.Since = date.strftime("%Y-%m-%d")
        c.Until = (date + datetime.timedelta(days=1)).strftime("%Y-%m-%d")
        twint.run.Search(c)
```
